# Put matched_reblogs_nonreblogs into feature tables for running WebSci experiments

```python
# Load WebSci data
import os
import pandas as pd

websci_dirpath = '/data/websci2020_tumblr_identity/websci2020_submission/'
features = {}

for fold in ['train', 'test']:
    features[fold] = {}
    for kind in ['reblog', 'nonreblog']:
        fpath = os.path.join(websci_dirpath, fold, 'feature_tables', f'{kind}_features.csv')
        features[fold][kind] = pd.read_csv(fpath)

print(features['train']['reblog'].columns)
features['train']['reblog']
```

```python
# Load matched_reblogs_nonreblogs
matched_reblogs_nonreblogs_fpath = '/data/tumblr_community_identity/dataset114k/matched_reblogs_nonreblogs_dataset114k_annotated.csv'
matched = pd.read_csv(matched_reblogs_nonreblogs_fpath, sep='\t')
print(matched.columns)
```

```python
# Fix for string-lists
for col in ['post_tags_reblog', 'post_tags_nonreblog']:
#     matched[col] = matched[f'{col}_str'].str.replace('[', "['")
#     matched[col] = matched[col].str.replace(", ", "', '")
#     matched[col] = matched[col].str.replace("]", "']")
#     matched[col] = matched[col].replace({"['']": "[]"})
#     matched[col]
    matched[col] = matched[f'{col}_str'].str.replace(r'[\[\]]', '').str.split(r',\s+')
    matched[col] = matched[col].map(lambda x: [] if x==[''] else x)
    for row in matched.loc[matched[col].isnull(), col].index:
        matched.at[row, col] = []

categories = [
    'age',
    'ethnicity/nationality',
    'fandoms',
    'gender',
    'gender/sexuality',
    'interests',
    'location',
    'personality type',
    'pronouns',
    'relationship status',
    'roleplay',
    'roleplay/fandoms',
    'sexual orientation',
    'weight',
    'zodiac',
]
```

```python
matched[col]
```

```python
# Select, rename columns in reblog set

column_match_reblog = { # matched_reblog_nonreblog col : reblog_features col
    'post_id_reblog': 'post_id', 
    'tumblog_id_follower_reblog': 'tumblog_id_follower',
    'tumblog_id_followee_reblog': 'tumblog_id_followee',
    'post_tags_reblog': 'post_tags',
    'post_type_reblog': 'post_type', 
    'post_note_count_reblog': 'post_note_count', 
    'processed_tumblr_blog_description_follower_reblog': 'processed_blog_description_follower',
    'processed_tumblr_blog_description_followee_reblog': 'processed_blog_description_followee', 
}

for cat in categories:
    column_match_reblog[f'{cat}_terms_follower'] = f'{cat}_terms_follower'
    column_match_reblog[f'{cat}_terms_followee_reblog'] = f'{cat}_terms_followee'
    
matched_reblogs = matched.loc[:, list(column_match_reblog.keys())]
matched_reblogs.rename(columns=column_match_reblog, inplace=True)

# Reorder to match reblog feature table
matched_reblogs = matched_reblogs.loc[:, features['train']['reblog'].columns.tolist()]
print(all(matched_reblogs.columns == features['train']['reblog'].columns))

# Save out reblogs
matched_reblogs.to_csv('/data/tumblr_community_identity/dataset114k/feature_tables/reblog_features.csv', index=False)
```

```python
# Select, rename columns in nonreblog set
column_match_nonreblog = { # matched_reblog_nonreblog col : reblog_features col
    'post_id_nonreblog': 'post_id', 
    'tumblog_id_follower_reblog': 'tumblog_id_follower',
    'tumblog_id_followee_nonreblog': 'tumblog_id_followee',
    'post_tags_nonreblog': 'post_tags',
    'post_type_nonreblog': 'post_type', 
    'post_note_count_nonreblog': 'post_note_count', 
    'processed_tumblr_blog_description_follower_reblog': 'processed_blog_description_follower',
    'processed_tumblr_blog_description_followee_nonreblog': 'processed_blog_description_followee', 
}

for cat in categories:
    column_match_nonreblog[f'{cat}_terms_follower'] = f'{cat}_terms_follower'
    column_match_nonreblog[f'{cat}_terms_followee_nonreblog'] = f'{cat}_terms_followee'
    
matched_nonreblogs = matched.loc[:, list(column_match_nonreblog.keys())]
# [col for col in column_match_nonreblog if not col in matched]
matched_nonreblogs.rename(columns=column_match_nonreblog, inplace=True)

# Reorder to match nonreblog feature table
matched_nonreblogs = matched_nonreblogs.loc[:, features['train']['nonreblog'].columns.tolist()]
print(all(matched_nonreblogs.columns == features['train']['nonreblog'].columns))

# Save out nonreblogs
matched_nonreblogs.to_csv('/data/tumblr_community_identity/dataset114k/feature_tables/nonreblog_features.csv', index=False)
```

```python
# Generate random labels
import numpy as np

labels = [0] * (int(len(matched_reblogs)/2)) + [1] * (int(len(matched_reblogs)/2))
np.random.shuffle(labels)
print(len(labels))
print(labels.count(0))
print(labels.count(1))
```

```python
# Save ranking labels
out_dirpath = '/data/tumblr_community_identity/dataset114k/feature_tables/'
labels_fpath = os.path.join(out_dirpath, 'ranking_labels.csv')
with open(labels_fpath, 'w') as f:
    f.write('ranking_label\n')
    for label in labels:
        f.write(f'{label}\n')
```

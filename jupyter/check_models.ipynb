```python
# From post hashtag PCA
import numpy as np

def feats_for_factors(feature_names, pca, n_factors=20, n_feats=40):
    top = np.flip(np.argsort(pca.components_)[:n_factors, -1*n_feats:], axis=1)
    vec = np.vectorize(lambda x: feature_names[x])
    return vec(top)
factor_loadings = feats_for_factors(vec.get_feature_names(), svd, n_factors=50, n_feats=10)
factor_loadings
```

## Look at informative features

```python
selected = selected_feats + ['post_note_count', 'post_type_feat']
```

```python
import pdb
import pandas as pd
from operator import itemgetter
pd.set_option('display.max_rows', None)

def get_informative_features(featnames, model, n=10000):
# def get_informative_features(features_vectorizer, model, n=10000):
    """ Examine informative features from machine learning models. """

#     feats_index2name = {v: k for k, v in features_vectorizer.vocabulary_.items()}
#     feats_index2name = {v: k for k, v in featnames}
#     feature_weights = model.coef_[0][:len(featnames)]
    feature_weights = model.coef_[0]

    top_indices = np.argsort(feature_weights)[-1*n:]
    top_weights = np.sort(feature_weights)[-1*n:]
    bottom_indices = np.argsort(feature_weights)[:n]
    bottom_weights = np.sort(feature_weights)[:n]

    lines = [] # to sort and print

    for i, (j, w) in enumerate(zip(reversed(top_indices), reversed(top_weights))):
#         feature_name = feats_index2name[j]
        feature_name = featnames[j]
        lines.append([i+1, feature_name, w, abs(w)])

    for i, (j, w) in enumerate(zip(bottom_indices, bottom_weights)):
#         feature_name = feats_index2name[j]
        feature_name = featnames[j]
        lines.append([(i+1)*-1, feature_name, w, abs(w)])

#     lines = list(reversed(sorted(lines, key=itemgetter(3))))
    return pd.DataFrame(lines, columns = ['rank', 'feature', 'weight', 'absolute_weight']).sort_values('weight', ascending=False)

# get_informative_features(selected_feats, model, n=5)
df = get_informative_features(selected, model, n=10)
# get_informative_features(vec, model, n=20)
df[['feature', 'weight']]
```
